```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel("Y Lable")
plt.title("First cuve")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Oriented Object
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes.set_title('Larger plot')
axes2.plot(x,y)
axes2.set_title('Smaller plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('Plot 1')
axes[1].plot(x,y)
axes[1].set_title('Plot 2')


plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(x,y)
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=1,alpha=0.5,linestyle='-',marker='*',markersize=20)
```

# Exercices

```python
import numpy as np
x = np.arange(0,100)
y =x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,z,color='green')
ax1.set_xlabel('x')
ax1.set_ylabel('z')

ax2.plot(x,y,color='blue')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('ZOOM')
```

#### fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,color='blue',ls='--',lw=3)
axes[1].plot(x,z,color='red',ls='-',lw=3)

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(8,2))
axes[0].plot(x,y,color='blue',ls='-',lw=3)
axes[1].plot(x,z,color='red',ls='--',lw=3)
```
